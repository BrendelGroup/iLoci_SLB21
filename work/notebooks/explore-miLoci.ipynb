```python
from __future__ import division
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import pandas
import re
import seaborn
```

```python
def parse_seqreg(entry):
    seqregmatch = re.search('##sequence-region\s+(\S+)\s+(\d+)\s+(\d+)', entry)
    assert seqregmatch
    seqid = seqregmatch.group(1)
    start = int(seqregmatch.group(2))
    assert start == 1
    end = int(seqregmatch.group(3))
    return seqid, start, end
```

```python
def gff3_to_hist(species, seqid, root='../data'):
    seqlength = 0
    filename = '{}/{}/{}.iloci.gff3'.format(root, species, species)
    with open(filename, 'r') as infile:
        for line in infile:
            if line.startswith('##sequence-region'):
                seq, start, end = parse_seqreg(line)
                if seq == seqid:
                    seqlength = end
    
    milocus_starts = list()
    filename = '{}/{}/{}.miloci.gff3'.format(root, species, species)
    with open(filename, 'r') as infile:
        for line in infile:
            if 'iLocus_type=miLocus' not in line:
                continue
            fields = line.split('\t')
            assert len(fields) == 9
            seq = fields[0]
            if seq != seqid:
                continue
            start = int(fields[3]) / seqlength
            milocus_starts.append(start)
    return milocus_starts
```

```python
seaborn.set_context({"figure.figsize": (12, 6)})
chromosomes = 'NC_000067.6 NC_000068.7 NC_000069.6 NC_000070.6 NC_000071.6 NC_000072.6 NC_000073.6 NC_000074.6 NC_000075.6 NC_000076.6 NC_000077.6 NC_000078.6 NC_000079.6 NC_000080.6 NC_000081.6 NC_000082.6 NC_000083.6 NC_000084.6 NC_000085.6 NC_000086.7 NC_000087.7'
for chrom in chromosomes.split(' '):
    mipos = gff3_to_hist('Mmus', chrom)
    _ = plt.hist(mipos, bins=75)
    _ = plt.title(chrom)
    _ = plt.show()
```
